```python
#combine all together！！！！
import json
#3.set pattern of people's activities
#Reference: A House of Dust From Implementations of early and well-known poetry generators¶ By Allison Parrish
#Original written in Fortran in 1967 by Alison Knowles and James Tenney. ELMCIP entry. More information. Watch Alison Knowles read from this piece.

import random
#set the array of inhabitant's names(I hope this poem can be based on a fictional community and can be replayed. 
#Therefore, I've designed only ten villager names, hoping that the audience will feel familiar with their existence 
#during the experience.)
verb_tokens = []
subjects = []
objects = []
with open('LifeEventDialog_valid.jsonl', 'r') as file:
    for line in file:
        LED = json.loads(line)

        for event_group in LED['events']:
            for event in event_group:
                if 'event_type' in event and 'participant' in event:
                    verb_token = event['event_type']['Verb']['tokens']
                    verb_tokens.append(verb_token)
                 
                    #print("Verb Token:", verb_tokens)
                    
                    subject = event['participant']['subject']['tokens']
                    subjects.append(subject)
                    #print("Subject:", subjects)
                    
                    if 'tokens' in event['participant']['object']:
                        object_tokens = event['participant']['object']['tokens']
                        objects.append(object_tokens)
                        #print("Object:", objects)
                    #else:
                        #print("Object: No tokens found")
                        
if verb_tokens:    
    random_verb = random.choice(verb_tokens)
    random_verb2 = random.choice(verb_tokens)
    #print("Selected Verb Token:", random_verb)
if subjects:
    random_subjects=random.choice(subjects)
    #print("Selected subjects Token:", random_subjects)
if objects:
    random_objects=random.choice(objects)
    #print("Selected subjects Token:", random_objects)
subjects1=["He","She","They"]
inhabitants = [
    'thieves',
    'dealers',
    'friends',
    'merchant',
    'lovers',
    'police',
    'vegetarians',
    "Eleanor",  
    "Oliver",  
    "Amelia",   
    "Jack",     
    "Charlotte",
    "Henry",    
    "Isabella", 
    "Liam",     
    "Sophia",   
    "Noah"      
]
locations = [
    "Under the oak of Maple Street Park",     
    "By the gate of Riverside Café",          
    "Between the stacks of Willow Library",  
    "Behind the walls of Oakwood School",     
    "At the crossroads of Meadow Lane Market",
    "Near the old well of Cedar Clinic",      
    "Under the lanterns of Pine Plaza",      
    "In the shadow of Elm Avenue Bakery",     
    "Along the banks of Sunset Bridge",       
    "At the foot of Lighthouse Point"        
]
separator = "-----"
stanza_count = 4
#2.set fixed sentences
bor="Oh~ What a boring birds' life！"
see="Let's peek at the human hive,"+ "\n"+"see what's keeping them alive!"

for i in range(stanza_count):

    line2_words = see.split()
    line3_words = (random.choice(inhabitants) + " " + random_verb + " " + random.choice(objects)).split()
    line4_words = random.choice(locations).split()
    line5_words = (random.choice(subjects1) + " " + random.choice(verb_tokens)).split()
    line6_words = ("But " + random_objects + " " + random.choice(verb_tokens)).split()
    line7_words = bor.split()


    def insert_separators(word_list):
        for _ in range(random.randint(1, len(word_list))):  # 插入1到单词数量之间的随机数量的分隔线
            word_list.insert(random.randint(0, len(word_list)), separator)
        return ' '.join(word_list)

    print(insert_separators(line2_words))
    print(insert_separators(line3_words))
    print(insert_separators(line4_words))
    print(insert_separators(line5_words))
    print(insert_separators(line6_words))
    print(insert_separators(line7_words))
    print("")  
```
